```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataframe = pd.read_csv("aviation-accidents.csv")
dataframe
```

```python
def clean_aviation_data(df):
    # Replace "unknown", "unknown date", "unknown country" with NaN
    unknown_values = ["unknown", "date unk.", "Unknown country"]
    df.replace(unknown_values, np.nan, inplace=True)

    # Convert 'fatalities' to numeric (forcing errors to NaN)
    df['fatalities'] = pd.to_numeric(df['fatalities'], errors='coerce')

    # Remove leading and trailing whitespace in text values
    df = df.map(lambda x: x.strip() if isinstance(x, str) else x)

    #  Show the number of missing values ​​after cleaning
    print("Missing values ​​after cleaning:")
    print(df.isnull().sum())

    return df

#  Example of use:
# df = clean_aviation_data(df)
```

```python
dataframe.dropna(inplace=True)
dataframe.reset_index(drop=True)
```

```python
def plot_accidents_by_country(df, top_n=10):
    country_counts = df['country'].value_counts().head(top_n)

    plt.figure(figsize=(12, 6))
    country_counts.plot(kind='bar', color='green')
    plt.xlabel("Nation")
    plt.ylabel("Number of accidents")
    plt.title(f"Top {top_n} Countries with the Most Plane Accidents")
    plt.xticks(rotation=45)
    plt.show()

    # plot_accidents_by_country(df)
```

```python
plot_accidents_by_country(dataframe)
```

```python


def plot_accident_trends(df):
    # Convert 'date' column to datetime, forcing NaT errors
    df['date'] = pd.to_datetime(df['date'], errors='coerce')

    # Remove any invalid dates (NaT converted to NaN)
    df.dropna(subset=['date'], inplace=True)

    # Create additional columns
    df['weekday'] = df['date'].dt.day_name()
    df['year'] = df['date'].dt.year

    # Accidents by day of the week
    weekday_counts = df['weekday'].value_counts()
    weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    weekday_counts = weekday_counts.reindex(weekday_order)

    # Pre- and Post-September 11 Incident Chart
    df['post_9_11'] = df['date'] >= pd.Timestamp('2001-09-11')
    yearly_counts = df.groupby(['year', 'post_9_11']).size().unstack()

    # Creating charts
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Accident chart by day of the week
    axes[0].bar(weekday_counts.index, weekday_counts.values, color='orange')
    axes[0].set_title("Accidents by day of the week")
    axes[0].set_xlabel("Day of the week")
    axes[0].set_ylabel("Number of accidents")
    axes[0].tick_params(axis='x', rotation=45)

    # Pre- and Post-September 11 Incident Chart
    if not yearly_counts.empty:
        yearly_counts.plot(kind='line', marker='o', ax=axes[1],color=['red', 'green'])
        axes[1].set_title("Pre and Post September 11, 2001 Plane Crashes")
        axes[1].set_xlabel("Year")
        axes[1].set_ylabel("Number of accidents")
        axes[1].legend(["Pre 11/09/2001", "Post 11/09/2001"])
    else:
        axes[1].text(0.5, 0.5, "No data available", ha='center', va='center', fontsize=12)
        axes[1].set_xticks([])
        axes[1].set_yticks([])

    plt.tight_layout()
    plt.show()

```

```python
plot_accident_trends(dataframe)
```

```python
def plot_safety_analysis(df):
    operator_safety = df['operator'].value_counts()
    safe_operators = operator_safety[operator_safety >= 20].tail(20)
    aircraft_fatalities = df.groupby('type')['fatalities'].sum().sort_values(ascending=False).head(10)

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

     # Safer Operators Chart
    axes[0].barh(safe_operators.index, safe_operators.values, color='skyblue')
    axes[0].set_title("Safer operators")
    axes[0].set_xlabel("Number of accidents")
    axes[0].set_ylabel("Operator")
    axes[0].set_xlim(0,50)



    # Chart of aircraft types with most deaths
    axes[1].barh(aircraft_fatalities.index, aircraft_fatalities.values, color='red')
    axes[1].set_title("Aircraft types with most deaths")
    axes[1].set_xlabel("Number of Fatalities")
    axes[1].set_ylabel("Aircraft types")


    plt.tight_layout()
    plt.show()
```

```python
plot_safety_analysis(dataframe)
```

```python
import geopandas as gpd

def plot_accidents_cartogram(df):
    # Download the Natural Earth dataset
    world = gpd.read_file("https://raw.githubusercontent.com/nvkelso/natural-earth-vector/master/geojson/ne_110m_admin_0_countries.geojson")

    country_counts = df['country'].value_counts().reset_index()
    country_counts.columns = ['country', 'accidents']

    # Correct country names to match geographic dataset
    country_corrections = {
        "USA": "United States of America",
        "U.K.": "United Kingdom",
    }
    country_counts['country'] = country_counts['country'].replace(country_corrections)

    world = world.merge(country_counts, left_on='NAME', right_on='country', how='left')
    world = world.infer_objects(copy=False)  # Converts objects to appropriate types
    world['accidents'] = world['accidents'].fillna(0)  # Fills NaN with 0 only for the 'accidents' column'

    fig, ax = plt.subplots(figsize=(15, 8))
    world.plot(column='accidents', cmap='Blues', linewidth=0.8, edgecolor='black', legend=True, ax=ax)
    ax.set_title("Number of Air Accidents by Country")

    plt.show()

def top_five_countries(df):
    top_countries = df['country'].value_counts().head(5)
    print("Top 5 Countries by Number of Accidents:")
    print(top_countries)
```

```python
plot_accidents_cartogram(dataframe)
top_five_countries(dataframe)
```
